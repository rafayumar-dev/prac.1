```python
import pandas as pd
import matplotlib.pyplot as plt

attendance = {
    'names': ['rafay', 'bilal', 'abubakar', 'zain', 'hussain', 'junaid'],
    'present': [24, 15, 30, 25, 28, 12]
}

at = pd.DataFrame(attendance)

bars = plt.bar(at['names'], at['present'], label="present")

plt.title("PMB-1 Attendance")
plt.xlabel("Names")
plt.ylabel("Days Present")
plt.legend()

#  ADD VALUES ON TOP OF BARS (IMPORTANT PART)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             str(height),
             ha='center', va='bottom')

plt.xticks(rotation=0)
plt.show()
```
